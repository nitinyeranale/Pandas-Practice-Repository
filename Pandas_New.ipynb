#### Combine two datasets ------------------

```python
data1 = pd.DataFrame({'x1': ['a','b','c'], 'x2': [11.432, 1.303, 99.906]})
data2 = pd.DataFrame({'x1': list('abd'), 'x3': [20.784,  33, 20.784]})
```

```python
data1
```

```python
data2
```

```python
data1.merge(data2, how='inner')
```

```python
data1.merge(data2,how='left' )
```

```python
pd.concat([data1,data2])
```

```python
d1=pd.DataFrame({"Name":["Nitin","Ramesh","Rutuja"],"Marks":[100,99,98]})
d2=pd.DataFrame({"Name":["Sonu","Shruti","Nitin"],"Marks":[89,98,100]})
```

```python
d1.merge(d2, on="Name")
```

```python
pd.concat([d1,d2])
```
